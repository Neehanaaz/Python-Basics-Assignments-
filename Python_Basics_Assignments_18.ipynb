You can create an OrderedDict called fancy from the same key-value pairs as in question 5 and print it out. The OrderedDict will maintain the order in which the key-value pairs are inserted. Here's an example code:

fancy = OrderedDict([('a', 1), ('b', 2), ('c', 3)])
print(fancy)
This will output the OrderedDict in the following form

OrderedDict([('a', 1), ('b', 2), ('c', 3)])
You'll notice that the output is the same as in question 5, but with the addition of the OrderedDict class name and parentheses.

And yes, the output of the fancy OrderedDict should be in the same order as the plain dictionary in question 5, with the key-value pairs 'a': 1, 'b': 2, and 'c': 3 appearing in that order. This is because we inserted the key-value pairs in the same order when creating the fancy OrderedDict.


7. Make a default dictionary called dict_of_lists and pass it the argument list. Make the list dict_of_lists['a'] and append the value 'something for a' to it in one assignment. Print dict_of_lists.

You can create a defaultdict called dict_of_lists with the default value set to a new empty list, and then append the value 'something for a' to the list associated with key 'a'. Here's the code:


from collections import defaultdict

dict_of_lists = defaultdict(list)
dict_of_lists['a'].append('something for a')
print(dict_of_lists['a'])
This will output the following:

['something for a']
In this code, we created a defaultdict with the default value set to an empty list using the list constructor as the default factory function. This means that when we access a key that does not exist in the dictionary, a new empty list will automatically be created as the default value for that key.

Next, we accessed the list associated with key 'a' and used the append method to add the value 'something for a' to the end of the list, all in one line of code.

Finally, we printed the value associated with key 'a' using dictionary indexing syntax to confirm that the value was added correctly.
