4) What is the purpose of a yield statement?
The yield statement is used in Python to define a generator function. The purpose of a yield statement is to yield a value from the generator function to the calling code, while also suspending the state of the generator function so that it can be resumed later.

When a yield statement is executed in a generator function, the value specified after the yield keyword is returned to the calling code as the next value in the sequence. The state of the generator function is then suspended, and the calling code can continue executing until it requests the next value from the generator object.

When the generator object is iterated over again (using the next() function or a for loop), the generator function is resumed from the point where it was suspended by the previous yield statement. The function then continues executing until it reaches the next yield statement, at which point it yields another value and suspends again.

This process continues until the generator function has yielded all the values in the sequence, at which point it raises a StopIteration exception to signal that there are no more values to be generated.

The yield statement is a powerful tool for generating sequences of values on-the-fly, especially when dealing with large or infinite sequences that would be impractical to generate all at once. By suspending the state of the generator function between yields, the yield statement allows the function to generate values lazily, as they are requested, making it more memory-efficient than other methods of generating sequences.
