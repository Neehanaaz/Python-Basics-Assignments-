1. What is the result of the code, and explain?


>>> X = 'iNeuron'
>>> def func():
print(X)


>>> func()

The code defines a variable X with the value 'iNeuron'. Then it defines a function called func that prints the value of X.

When the function func() is called, it will print the value of X, which is 'iNeuron'. Therefore, the output of running this code will be:

iNeuron





2. What is the result of the code, and explain?


>>> X = 'iNeuron'
>>> def func():
X = 'NI!'


>>> func()
>>> print(X)
The code defines a variable X with the value 'iNeuron'. It then defines a function called func that creates a new variable X with the value 'NI!' within the function's local scope.

When the function func() is called, it assigns the value 'NI!' to the variable X within its local scope. However, this does not affect the value of the global variable X which still has the value 'iNeuron'.

After calling the function, the code tries to print the value of X outside of the function. Since the value of X was not modified outside the function, it still has the value 'iNeuron'. Therefore, the output of running this code will be

iNeuron





3. What does this code print, and why?


>>> X = 'iNeuron'
>>> def func():

print(X)


>>> func()
>>> print(X)
The code defines a variable X with the value 'iNeuron'. It then defines a function called func that creates a new variable X with the value 'NI' within the function's local scope.

When the function func() is called, it assigns the value 'NI' to the variable X within its local scope. The function then prints the value of the local variable X, which is 'NI'.

After calling the function, the code tries to print the value of X outside of the function. Since the value of X was not modified outside the function, it still has the value 'iNeuron'. Therefore, the output of running this code will be:

NI
iNeuron
The first print statement within the function func() prints 'NI', which is the value of the local variable X within the function's scope. The second print statement outside the function prints 'iNeuron', which is the value of the global variable X that was not modified within the function.





4. What output does this code produce? Why?


>>> X = 'iNeuron'
>>> def func():
global X
X = 'NI'


>>> func()
>>> print(X)
The code defines a variable X with the value 'iNeuron'. It then defines a function called func that uses the global keyword to declare that it will modify the global variable X.

When the function func() is called, it modifies the value of the global variable X to 'NI'. Therefore, after calling the function, the value of the variable X is 'NI'.

The code then prints the value of X, which is 'NI'. Therefore, the output of running this code will be:

NI
Note that using the global keyword within a function allows the function to modify the value of a variable in the global scope. Without the global keyword, any assignment to a variable within the function creates a new variable in the local scope of the function.
